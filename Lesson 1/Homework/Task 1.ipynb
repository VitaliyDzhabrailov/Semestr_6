```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
grid_1 = np.array([1, 15])
A_1 = np.array([grid_1 ** 0, grid_1])
A_1 = A_1.transpose()
A_1
```

```python
b_1 = np.array([3.25, 0.25])
b_1
```

```python
f_1 = np.linalg.solve(A_1, b_1)
```

```python
grid_x = np.arange(0, 15, 0.1)
plt.plot(grid_x, f_1[0] + f_1[1] * grid_x)
plt.show()
```

```python
grid_2 = np.array([1, 8, 15])
A_2 = np.array([grid_2 ** 0, grid_2, grid_2 ** 2])
A_2 = A_2.transpose()
A_2
```

```python
b_2 = np.array([3.25, 2.25, 0.25])
b_2
```

```python
f_2 = np.linalg.solve(A_2, b_2)
```

```python
grid_x = np.arange(0, 15, 0.1)
plt.plot(grid_x, f_2[0] + f_2[1] * grid_x + f_2[2] * grid_x ** 2)
plt.show()
```

```python
grid_3 = np.array([1, 4, 8, 15])
A_3 = np.array([grid_3 ** 0, grid_3, grid_3 ** 2, grid_3 ** 3])
A_3 = A_3.transpose()
A_3
```

```python
b_3 = np.array([3.25, 1.75, 2.25, 0.25])
b_3
```

```python
f_3 = np.linalg.solve(A_3, b_3)
```

```python
grid_x = np.arange(0, 15, 0.1)
plt.plot(grid_x, f_3[0] + f_3[1] * grid_x + f_3[2] * grid_x ** 2 + f_3[3] * grid_x ** 3)
plt.show()
```

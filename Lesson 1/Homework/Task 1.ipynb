```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
grid_1 = np.array([1, 15])
A_1 = np.array([grid_1 ** 0, grid_1])
A_1 = A_1.transpose()
A_1
```

```python
def f(x):
    return math.sin(x / 5) * math.exp(x / 10) + 5 * math.exp(-x / 2)
b_1 = np.array([f(i) for i in grid_1])
b_1
```

```python
f_1 = np.linalg.solve(A_1, b_1)
```

```python
grid_x = np.arange(0, 15.01, 0.1)
plt.plot(grid_x, f_1[0] + f_1[1] * grid_x, label='Аппроксимация')
plt.plot(grid_x, np.array([f(i) for i in grid_x]), label='Оригинал')
plt.legend()
plt.show()
```

```python
grid_2 = np.array([1, 8, 15])
A_2 = np.array([grid_2 ** 0, grid_2, grid_2 ** 2])
A_2 = A_2.transpose()
A_2
```

```python
b_2 = np.array([f(i) for i in grid_2])
b_2
```

```python
f_2 = np.linalg.solve(A_2, b_2)
```

```python
grid_x = np.arange(0, 15.01, 0.1)
plt.plot(grid_x, f_2[0] + f_2[1] * grid_x + f_2[2] * grid_x ** 2, label='Аппроксимация')
plt.plot(grid_x, np.array([f(i) for i in grid_x]), label='Оригинал')
plt.legend()
plt.show()
```

```python
grid_3 = np.array([1, 4, 8, 15])
A_3 = np.array([grid_3 ** 0, grid_3, grid_3 ** 2, grid_3 ** 3])
A_3 = A_3.transpose()
A_3
```

```python
b_3 = np.array([f(i) for i in grid_3])
b_3
```

```python
f_3 = np.linalg.solve(A_3, b_3)
```

```python
grid_x = np.arange(0, 15.01, 0.1)
plt.plot(grid_x, f_3[0] + f_3[1] * grid_x + f_3[2] * grid_x ** 2 + f_3[3] * grid_x ** 3, label='Аппроксимация')
plt.plot(grid_x, np.array([f(i) for i in grid_x]), label='Оригинал')
plt.legend()
plt.show()
```

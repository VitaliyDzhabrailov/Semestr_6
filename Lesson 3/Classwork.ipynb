```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 10, 20)
y = 2 * np.sin(x)
```

```python
plt.rcParams('')
```

```python
fig, ax = plt.subplots(figsize=(8, 5))
ax.plot(x, y, color='purple', lw=5, marker='v', markersize=20, label='y=2sin(x)')
ax.plot(x, np.cos(y))
plt.legend(loc='lower left', title='Title')
plt.minorticks_on()
plt.grid(which='both')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8, 10), nrows=2, ncols=1)
ax[0].plot(x, y, color='purple', lw=5, marker='v', markersize=20, label='y=2sin(x)')
ax[1].plot(x, np.cos(y))
ax[0].set_xticklabels([])
ax[0].legend(loc='lower left', title='Title')
ax[0].minorticks_on()
ax[0].grid(which='both')
plt.show()
```

```python
plt.scatter(x, y, s=400)
plt.show()
```

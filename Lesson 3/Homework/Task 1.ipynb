```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv("flight_delays.csv")
data
```

Доля всех задержек ко всем вылетам

```python
data = data.assign(score=lambda x: "ALL")
fig, ax = plt.subplots(figsize =(4, 7))
ax.hist([data[data.dep_delayed_15min == 'Y'].dep_delayed_15min, data.score], bins=1)
labels = ['With delay', 'All flights']
rects = ax.patches
for rect, label in zip(rects, labels):
    height = rect.get_height()
    ax.text(rect.get_x() + rect.get_width() / 2, height+0.01, label, ha='center', va='bottom')
ax.set_ylabel("Number of flights")
plt.show()
```

В какие времена года чаще всего происходят задержки рейсов

```python
data = pd.read_csv("flight_delays.csv")
data
```

```python
data_season = data.assign(score=1)
data_season = data_season[data_season.dep_delayed_15min == 'Y'].groupby(by='Month').count()
data_season
```

```python
fig, ax = plt.subplots(figsize =(10, 7))
list1 = [data_season.loc['c-12'].score + data_season.loc['c-1'].score + data_season.loc['c-2'].score,
         data_season.loc['c-3'].score + data_season.loc['c-4'].score + data_season.loc['c-5'].score,
         data_season.loc['c-6'].score + data_season.loc['c-7'].score + data_season.loc['c-8'].score,
         data_season.loc['c-9'].score + data_season.loc['c-10'].score + data_season.loc['c-11'].score]

ax.bar(['winter', 'spring', 'summer', 'fall'], list1)
ax.set_ylabel("Number of delays")
ax.set_xlabel("Season")
plt.show()
```

Топ 10 самых хороших перевозчиков, которые реже всего задерживают свои рейсы

```python
data = pd.read_csv("flight_delays.csv")
data
```

```python
data_carrier = data.assign(score=1)
data_carrier = data_carrier[data_carrier.dep_delayed_15min == 'Y'].groupby(by='UniqueCarrier', as_index=False).count()
data_carrier.sort_values(by='score', ascending=False, inplace=True, ignore_index=True)
data_carrier
```

```python
fig, ax = plt.subplots(figsize =(10, 7))
ax.bar(data_carrier.UniqueCarrier[0:10], data_carrier.score[0:10])
ax.set_ylabel("Number of delays")
ax.set_xlabel("Carrier")
plt.show()
```

топ 10 самых безответственных аэропортов, в которых чаще всего происходят задержки

```python
data = pd.read_csv("flight_delays.csv")
data
```

```python
data_origin = data.assign(score=1)
data_origin = data_origin[data_origin.dep_delayed_15min == 'Y'].groupby(by='Origin', as_index=False).count()
data_origin.sort_values(by='score', ascending=False, inplace=True, ignore_index=True)
data_origin
```

```python
fig, ax = plt.subplots(figsize =(10, 7))
ax.bar(data_origin.Origin[0:10], data_origin.score[0:10])
ax.set_ylabel("Number of delays")
ax.set_xlabel("Origin")
plt.show()
```

Необычная зависимость количества задержек от имеющихся данных

```python
data = pd.read_csv("flight_delays.csv")
data
```

```python
data_kek = data.assign(velocity=lambda x: x.Distance / x.DepTime)
data_kek
```

```python
fig, ax = plt.subplots(figsize =(15, 7))
ax.hist(data_kek[data_kek.dep_delayed_15min == 'Y'].velocity, bins=25000)
ax.set_ylabel("Number of delays")
ax.set_xlabel("\"Velocity\"")
ax.set_xlim(0, 2)
plt.show()
```
